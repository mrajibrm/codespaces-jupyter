```python
from reportlab.lib.pagesizes import letter, landscape
from reportlab.pdfgen import canvas
from reportlab.lib.utils import ImageReader
from reportlab.pdfbase.pdfmetrics import stringWidth
import matplotlib.pyplot as plt
import re
import io
```

## Working

```python

def create_pdf_with_latex(df, filename="output_with_tex.pdf"):
    
    # Set the canvas
    c = canvas.Canvas(filename, pagesize=landscape(letter))
    width, height = landscape(letter)

    # Define Layout
    left_margin = 50
    top_margin = height - 50
    line_spacing = 20
    max_line_width = 575

    # Helper function to wrap text within the canvas
    def wrap_text(text, max_width):
        words = text.split()
        lines = []
        current_line = words[0]
        for word in words[1:]:
            if stringWidth(current_line + " " + word, 'Helvetica', 10) < max_width:
                current_line += " " + word
            else:
                lines.append(current_line)
                current_line = word
        lines.append(current_line)
        return lines
    
    # Helper function to render LaTeX eqn
    def render_latex_eqn(latex_sring):
        fig, ax = plt.subplots()
        ax.text(0.5, 0.5, f"${latex_sring}$", fontsize=12, ha='center', va='center')
        ax.axis('off')
        buffer = io.BytesIO()
        fig.savefig(buffer, format='png', bbox_inches='tight', pad_inches=0)
        plt.close(fig)
        buffer.seek(0)
        return ImageReader(buffer)
    
    # Loop through each row in the dataframe
    for index, row in df.iterrows():
        y_position = top_margin

        # For Each Column name and value in the row
        for col, value in row.items():
            text = f"{col}:{value}"
            lines = wrap_text(text, max_line_width)

            for line in lines:
                # Check for Latex Eqn
                latex_match = re.findall(r'\$(.*?)\$', line)

                if latex_match:
                    for eqn in latex_match:
                        eqn_pos = line.find(f"${eqn}$")
                        prefix_text = line[:eqn_pos]

                        # Draw text upto the equaion
                        c.drawString(left_margin, y_position, prefix_text)

                        # Render and place LaTeX Eqn Image
                        latex_img = render_latex_eqn(eqn)
                        c.drawImage(latex_img, left_margin + stringWidth(prefix_text, 'Helvetica', 10), y_position - 10, width = 50, height = 20)

                        # Continue after equation
                        remaining_text = line[eqn_pos + len(eqn) + 2:]
                        next_pos = left_margin + stringWidth(prefix_text, 'Helvetica', 10) + 55
                        c.drawString(next_pos, y_position, remaining_text)

                else:
                    # Draw text without LaTeX EQn
                    c.drawString(left_margin, y_position, line)

                y_position -= line_spacing # Move down for nex line

            y_position -= line_spacing # Extra space after each row
        c.showPage() # For new page for each row

    c.save()
    

```

```python
create_pdf_with_latex(df_cleaned)
```

```python
import pandas as pd
from reportlab.lib.pagesizes import letter, landscape
from reportlab.platypus import SimpleDocTemplate, Paragraph, Image, Spacer
from reportlab.lib.styles import getSampleStyleSheet
import matplotlib.pyplot as plt
import re
import io

# Sample DataFrame with text, equations, and longer descriptions
# data = {
#     'Name': ['Alice', 'Bob', 'Charlie'],
#     'Age': [24, 27, 22],
#     'Country': ['USA', 'Canada', 'UK'],
#     'Description': [
#         'Alice is an engineer specializing in AI. Equation: $E = mc^2$',
#         'Bob is a neurologist. Research: $\\alpha + \\beta = \\gamma$',
#         'Charlie is an artist who works with multimedia. Formula: $f(x) = ax^2 + bx + c$'
#     ]
# }
# df = pd.DataFrame(data)

# Function to render LaTeX equations to image in BytesIO format
def render_latex_to_image(latex_string):
    fig, ax = plt.subplots()
    ax.text(0.5, 0.5, f"${latex_string}$", fontsize=12, ha='center', va='center')
    ax.axis('off')
    buf = io.BytesIO()
    fig.savefig(buf, format='png', bbox_inches='tight', pad_inches=0)
    plt.close(fig)
    buf.seek(0)
    return buf

# PDF generation function
def create_pdf_with_simpledoc(df, filename="output_with_latex_simpledoc.pdf"):
    # Set up document
    doc = SimpleDocTemplate(filename, pagesize=landscape(letter))
    styles = getSampleStyleSheet()
    content = []

    # Spacer to create space between lines
    spacer = Spacer(1, 12)

    # Loop through each row in the DataFrame
    for _, row in df.iterrows():
        for col, value in row.items():
            text = f"{col}: {value}"

            # Process text and look for LaTeX equations
            latex_matches = re.findall(r'\$(.*?)\$', text)
            if latex_matches:
                split_text = re.split(r'\$(.*?)\$', text)  # Split text by LaTeX parts
                for i, segment in enumerate(split_text):
                    if i % 2 == 0:
                        # Add non-LaTeX text
                        if segment.strip():  # Only add non-empty segments
                            content.append(Paragraph(segment, styles['BodyText']))
                    else:
                        # Render LaTeX equation to image in BytesIO format
                        latex_img_stream = render_latex_to_image(segment)
                        img = Image(latex_img_stream, width=50, height=20)  # adjust size as needed
                        content.append(img)
                    content.append(spacer)  # add space after each part
            else:
                # If no LaTeX, add text directly
                content.append(Paragraph(text, styles['BodyText']))
                content.append(spacer)

        # Add a page break after each row
        content.append(Spacer(1, 24))
        doc.build(content)
        content.clear()  # Clear content to start a new page for each row

# Generate PDF
create_pdf_with_simpledoc(df_cleaned)
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from reportlab.pdfgen import canvas
from reportlab.lib.pagesizes import letter, landscape
import re
import io

# Create a directory for LaTeX images
latex_img_dir = "latex_images"
os.makedirs(latex_img_dir, exist_ok=True)

# Sample DataFrame with text, equations, and longer descriptions
# data = {
#     'Name': ['Alice', 'Bob', 'Charlie'],
#     'Age': [24, 27, 22],
#     'Country': ['USA', 'Canada', 'UK'],
#     'Description': [
#         'Alice is an engineer specializing in AI. Equation: $E = mc^2$',
#         'Bob is a neurologist. Research: $\\alpha + \\beta = \\gamma$',
#         'Charlie is an artist who works with multimedia. Formula: $f(x) = ax^2 + bx + c$'
#     ]
# }
# df = pd.DataFrame(data)

# Function to render LaTeX equations to images and save to directory
def render_latex_to_image(latex_string, filename):
    fig, ax = plt.subplots()
    ax.text(0.5, 0.5, f"${latex_string}$", fontsize=12, ha='center', va='center')
    ax.axis('off')
    fig.savefig(filename, format='png', bbox_inches='tight', pad_inches=0)
    plt.close(fig)

# PDF generation function with LaTeX image insertion
def create_pdf_from_dataframe(df, filename="output.pdf"):
    c = canvas.Canvas(filename, pagesize=landscape(letter))
    width, height = landscape(letter)

    # Layout settings
    left_margin = 50
    top_margin = height - 50
    line_spacing = 20

    for index, row in df.iterrows():
        y_position = top_margin

        for col, value in row.items():
            text = f"{col}: {value}"

            # Process text and look for LaTeX equations
            latex_matches = re.findall(r'\$(.*?)\$', text)
            if latex_matches:
                split_text = re.split(r'\$(.*?)\$', text)  # Split text by LaTeX parts
                x_position = left_margin  # Start x-position for each row

                for i, segment in enumerate(split_text):
                    if i % 2 == 0:
                        # Draw non-LaTeX text
                        c.drawString(x_position, y_position, segment)
                        x_position += c.stringWidth(segment, "Helvetica", 10)
                    else:
                        # Render and insert LaTeX image
                        latex_img_path = os.path.join(latex_img_dir, f"latex_{index}_{i}.png")
                        render_latex_to_image(segment, latex_img_path)
                        c.drawImage(latex_img_path, x_position, y_position - 10, width=50, height=20)
                        x_position += 50  # Adjust x-position based on image width

            else:
                # Draw plain text if no LaTeX
                c.drawString(left_margin, y_position, text)

            y_position -= line_spacing  # Move to the next line in the row

        c.showPage()  # Start a new page for each row

    c.save()  # Save the PDF

# Generate PDF with LaTeX-rendered images
create_pdf_from_dataframe(df_cleaned)
```
